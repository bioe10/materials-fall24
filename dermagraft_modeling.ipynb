# Oxygen transfer through a cell dermagraft
## Objective: Model the concentration of oxygen through a cellular layer using 1D Reaction-Diffusion

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
from ipywidgets import interact, FloatSlider
```

```python
# Define the parameters
R = 2.5e-18 # mol/cm/s
R = R*100 # mol/m/s
L = 300e-6 # m
D = 1e-9 # m2/s
C0 = 0.15 # mol/m3
```

```python
def scientific_notation(x, pos):
    # Convert the number to scientific notation with LaTeX-style formatting
    return fr"${x:.2e}$"
```

```python
# Create an array of x values
x = np.linspace(-L, L, 100)  # Include both positive and negative x values

# Calculate the corresponding y values using the equation
y = 1 + (R * L**2) / (D * C0) * (x**2 / L**2 - 1)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('c(x)/c0')

# Format the y-axis using the custom scientific notation function
plt.gca().yaxis.set_major_formatter(scientific_notation)

# Set the y-axis limits to display from 0 to 1
plt.ylim(-.1, 1.1)

plt.title('Plot of c(x)/c0')
plt.grid(True)
plt.show()
```

### Create a slide bar to understand the effect of the dimensionless number on diffusion

```python
# Define the plotting function
def plot_equation(dimnum):
    y = 1 + dimnum * (x**2 / L**2 - 1)

    # Create the plot
    plt.figure(figsize=(8, 6))
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Plot of y = 1 + dimnum * (x^2 / L^2 - 1)')
    plt.ylim(-.1, 1.1)
    plt.grid(True)
    plt.show()

# Create a slider widget for dimnum
dimnum_slider = FloatSlider(value=1.0, min=0.001, max=10.0, step=0.01, description='dimnum')

# Create an interactive plot
interact(plot_equation, dimnum=dimnum_slider)
```

```python
# Define a list of dimnum values from 0 to 1 by 0.1 increments
dimnum_values = np.arange(0, 1.1, 0.1)

# Create a plot for each dimnum value
plt.figure(figsize=(10, 6))
for dimnum in dimnum_values:
    y = 1 + dimnum * (x**2 / L**2 - 1)
    plt.plot(x, y, label=f'dimnum={dimnum:.1f}')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = 1 + dimnum * (x^2 / L^2 - 1) for Different dimnum Values')
plt.grid(True)
plt.legend()
plt.show()
```
